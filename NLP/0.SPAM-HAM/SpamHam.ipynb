```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
```

```python
df=pd.read_csv('email.csv',sep='\t',names=['label','message'])
```

```python
df.loc[df["label"]=="spam","label"]=1
df.loc[df["label"]=="ham","label"]=0
```

```python
x=df["message"]
y=df["label"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
cv=CountVectorizer()
X_train_cv=cv.fit_transform(X_train)
```

```python
X_train_cv.toarray()
```

```python
cv.vocabulary_
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import nltk
nltk.download('punkt')
nltk.download('stopwords')
stemmer = PorterStemmer()
def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [token.lower() for token in tokens if token.isalpha()]
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    tokens = [stemmer.stem(token) for token in tokens]

    return tokens

X_train = X_train.apply(preprocess_text)
X_test = X_test.apply(preprocess_text)
```

```python
# Convert the list of tokens back into a string for each document
X_train = X_train.apply(' '.join)
X_test = X_test.apply(' '.join)

tf=TfidfVectorizer()
X_train_tf=tf.fit_transform(X_train)
X_test_tf=tf.transform(X_test)
```

```python
X_train=X_train_tf.toarray()
X_test=X_test_tf.toarray()
```

```python
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
nvb=MultinomialNB()
nvb.fit(X_train,y_train)
```

```python
y_pred=nvb.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam


model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))


model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])


model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)


 accuracy = model.evaluate(X_test, y_test)
print('Accuracy:', accuracy)
```
